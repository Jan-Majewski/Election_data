```python
def substring_after(s, delim):
    return s.partition(delim)[2]

def substring_before(s, delim):
    return s.partition(delim)[0]
```

```python
def trim_powiat(s):
    s_t1=substring_after(s,":")
    s_t2=substring_after(s_t1,":")
    s_t1=substring_before(s_t1,"oraz")
    if  s_t2=="":
        return (s_t1)
    else:
        return (s_t1+","+s_t2)
```
